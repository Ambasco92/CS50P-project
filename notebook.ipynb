# Date Utility Library (Python)
This project is a Python-based date utility library that provides several functions for validating, formatting, manipulating, and analyzing dates. The project accepts dates in the format YYYY-MM-DD and allows users to extract different components of a date, perform calculations, and convert dates into different readable formats.

The goal of the project is to implement functionality similar to date functions commonly found in SQL databases and spreadsheet software, but entirely in Python. The program allows users to work with dates without needing complex datetime logic.

This project demonstrates the use of Python functions, regular expressions, dictionaries, error handling, and built-in modules to build a reusable date-processing tool.

## Extracting date parts


`day(date_str)`

Accepts date as a string and extracts in the format `YYYY-MM-DD` and returns the day section `DD` as an `int`.

Examples:

```python
day("2026-03-12")
```

```python
day("2016-03-05")
```

`month(date_str)`

Accepts date as a string and extracts in the format `YYYY-MM-DD` and returns the month section `MM` as an `int`.

Examples:

```python
month("2026-03-12")
```

```python
month("2026-07-12")
```

`year(date_str)`

Accepts date as a string and extracts in the format `YYYY-MM-DD` and returns the year section `YYYY` as an `int`.

Examples:

```python
year("2026-03-12")
```

```python
year("2002-10-24")
```

`week(date_str)`

Accepts date as a string in the format `YYYY-MM-DD` and returns what week of the year the date belongs to as an `int`.

Between `week 1` to `week 52`.

Examples:

```python
week("2026-03-12")
```

```python
week("1990-07-24")
```

`week_day(date_str)`

Accepts date as a string in the format `YYYY-MM-DD` and returns what day of the week the date belongs to as an `int`.

Output:
| Day | Number |
|:--------:|:-------:|
| Monday | 0 |
| Tuesday | 1 |
| Wednesday | 2 |
| Thursday | 3 |
| Friday | 4 |
| Saturday | 5 |
| Sunday | 6 |

Examples:

```python
week_day("2026-03-12")
```

```python
week_day("1990-07-24")
```

`quarter(date_str)`

Accepts string date in the format `YYYY-MM-DD` and returns what quarter of the year that date belongs to as an `int` .

Examples:

```python
quarter("2026-10-24")
```

```python
quarter("1990-07-24")
```

`day_of_week(date_str)`

Accepts date as a string in the format `YYYY-MM-DD` and returns what day of the week the date belongs to as an `str`.

Examples:

```python
day_of_week("2026-03-12")
```

```python
day_of_week("2010-07-12")
```

`day_of_year(date_str)`

Accepts date as a string in the format `YYYY-MM-DD` and returns what day of the year that day is, as an `int`.

Between `day 1` to `day 365`, or between `day 1` to `day 366` for a leap year.

Examples:

```python
day_of_year("2021-12-31")
```

```python
day_of_year("2026-03-12")
```

### Date Validation

`is_date(date_str)`

Examples:

```python
is_date("2026-03-12")
```

```python
is_date("2016")
```

```python
is_date("2020-02-29")
```

`date_part(date_str, part='year')`

Returns a integer representation of a specific part of a date.

Args:

    date_str: A string representing a date in the format YYYY-MM-DD.

    part: One of ['day', 'month', 'year', 'week', 'week_day', 'quarter'].

Returns:

    int: The requested part of the date.

Raises:

    TypeError: If the date is invalid or the part is unknown.

Examples:

```python
date_part("2026-03-12")
```

```python
date_part("2026-03-12", part='month')
```

```python
date_part("2026-03-12", part='week')
```

### Date Truncation

Raises:

    TypeError: If `arg` is not 'day', 'month', or 'year'.

```python
date_trunc("2026-03-12")
```

### Date Formatting

`date_format(date_str, d_format='dd')`

Returns a string representation of the formatted version of a date.

Args:

    date_str: A string representing a date in the format YYYY-MM-DD.


Returns:

    str: The requested part of the date.

Raises:

    TypeError: If the date is invalid or the d_format is unknown.

Outputs:
| Input | d_format | Output |
|--------|-------|------|
| 2025-07-10 | `ddd` | Thu |
| 2025-07-10 | `dddd` | Thursday |
| 2020-10-12 | `dd-mm-yy` | 12-10-26 |
| 2026-02-12 | `dd-mmm-yyyy` | 12 Feb, 2026 |
| 2026-02-12 | `dd-mmmm-yyyy` | 12 February, 2026 |
...

Example:

```python
date_format("2025-07-10", d_format='dddd')
```

```python
date_format("2026-03-12", d_format='dd-mmm-yyyy')
```

```python
date_format("2026-03-12", d_format='dd-mmmm-yyyy')
```

```python
date_format("2026-03-12", d_format='dd-mm-yyyy')
```

### Date Addition

`date_add(date_str, interval, unit)`

Returns a string result of the addition to the date.

Args:

    date_str: A string representing a date in the format YYYY-MM-DD.
    interval: The interger interval amount you want to add to the date.
    unit: The unit of the interval to be added to date, ['days', 'weeks', 'months', 'years'].


Returns:

    str: The result of the addition to the date.

Raises:

    TypeError: If the date is invalid or unit not in list ['days', 'weeks', 'months', 'years'].

Examples:

```python
print(date_add("2026-03-12", interval= 10, unit='days'))
```

```python
print(date_add("2026-03-12", interval= 5, unit='weeks'))
```

```python
print(date_add("2026-03-12", interval= 10, unit='months'))
```

```python
print(date_add("2026-03-12", interval= 10, unit='years'))
```

### Date Subtraction

`date_diff(start_date, end_date, unit='days')`

Returns a string result of the subtraction of two dates.

Args:

    start_date: A string representing a date in the format YYYY-MM-DD.
    end_date: A string representing a date in the format YYYY-MM-DD.
    unit: The unit of substraction of the two dates, ['days', 'weeks', 'months', 'years'].


Returns:

    int: The result of the addition to the date.

Raises:

    TypeError: If one of the date is invalid or unit not in list ['days', 'weeks', 'months', 'years'].

Examples:

```python
date_diff("2026-03-12", "2027-03-13", unit='weeks')
```

```python
date_diff("2026-03-12", "2027-03-13", unit='years')
```

```python
date_diff("2026-03-12", "2027-03-13", unit='days')
```
